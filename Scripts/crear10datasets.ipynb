```python
import pandas as pd

# 1. Cargar dataset original
ruta = "dataset_base_features.csv"
df = pd.read_csv(ruta)

# 2. Agrupar por (product_id, customer_id) y sumar tn
agrupado = df.groupby(['product_id', 'customer_id'], as_index=False)['tn'].sum()

# 3. Ordenar de mayor a menor
agrupado = agrupado.sort_values(by='tn', ascending=False).reset_index(drop=True)

# 4. Calcular tamaño de cada decil
n = len(agrupado)
tam_decil = n // 10

# 5. Asignar deciles (1 = más tn, 10 = menos tn)
deciles = []
for i in range(10):
    start = i * tam_decil
    end = (i + 1) * tam_decil if i < 9 else n
    agrupado.loc[start:end, 'decil'] = i + 1

# 6. Merge con el dataset original para traer todas las filas
df_merged = df.merge(agrupado[['product_id', 'customer_id', 'decil']], on=['product_id', 'customer_id'])

# 7. Guardar cada decil como CSV
for i in range(1, 11):
    df_decil = df_merged[df_merged['decil'] == i]
    df_decil.to_csv(f"ds_{i}.csv", index=False)
```

```python
import pandas as pd

# 1. Cargar dataset
ruta = "dataset_base_features.csv"
df = pd.read_csv(ruta)

# 2. Agrupar por (product_id, customer_id) y sumar tn total
agrupado = df.groupby(['product_id', 'customer_id'], as_index=False)['tn'].sum()

# 3. Ordenar de mayor a menor
agrupado = agrupado.sort_values(by='tn', ascending=False).reset_index(drop=True)

# 4. Calcular tn total
tn_total = agrupado['tn'].sum()

# 5. Asignar deciles por masa acumulada
decil_col = []
acumulado = 0
decil_actual = 1
umbral = tn_total * 0.1

for idx, row in agrupado.iterrows():
    acumulado += row['tn']
    decil_col.append(decil_actual)
    if acumulado >= umbral * decil_actual and decil_actual < 10:
        decil_actual += 1

agrupado['decil'] = decil_col

# 6. Merge con el original
df_merged = df.merge(agrupado[['product_id', 'customer_id', 'decil']], on=['product_id', 'customer_id'])

# 7. Guardar datasets como dstn_1.csv, ..., dstn_10.csv
for i in range(1, 11):
    df_decil = df_merged[df_merged['decil'] == i]
    df_decil.to_csv(f"dstn_{i}.csv", index=False)

# 8. Mostrar suma de tn por decil
print("📊 Suma de tn por decil:")
print(df_merged.groupby('decil')['tn'].sum().round(2))
```

```python
import pandas as pd

# Reemplazar NaN por 0 en cada archivo
for i in range(1, 11):
    archivo = f"dstn_{i}.csv"
    df = pd.read_csv(archivo)
    df.fillna(0, inplace=True)
    df.to_csv(archivo, index=False)
    print(f"✅ Procesado: {archivo}")
```
