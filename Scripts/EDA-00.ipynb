```python
def reducirDT(df, verbose=True):
    inicio_mem = df.memory_usage(deep=True).sum() / 1024**2
    
    for col in df.select_dtypes(include=['int', 'float']).columns:
        col_min = df[col].min()
        col_max = df[col].max()
        if pd.api.types.is_integer_dtype(df[col]):
            if col_min >= -32768 and col_max <= 32767:
                df[col] = df[col].astype('int16')
            elif col_min >= -2147483648 and col_max <= 2147483647:
                df[col] = df[col].astype('int32')
            else:
                df[col] = df[col].astype('int64')
        else:
            df[col] = df[col].astype('float32')
    
    fin_mem = df.memory_usage(deep=True).sum() / 1024**2
    
    if verbose:
        ahorro = 100 * (inicio_mem - fin_mem) / inicio_mem
        print(f"ðŸ”§ Memoria reducida de {inicio_mem:.2f} MB a {fin_mem:.2f} MB ({ahorro:.1f}% de ahorro)")
    
    return df
```
